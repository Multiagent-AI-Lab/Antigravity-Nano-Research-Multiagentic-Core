```python
# ============================================================
#   SECCION 1: COMPLETA TU PROPUESTA DE PROYECTO
#   Rellena cada campo. Cuanto mas detalle, mejor retroalimentacion
#   recibiras del instructor y del validador automatico.
# ============================================================

mi_propuesta = {

    # --- Identificacion ---
    "nombre": "...",            # Tu nombre (o nombres si es en pareja)
    "fecha": "2026-...",        # Fecha de entrega de la propuesta

    # --- El problema ---
    "titulo": """
    [ESCRIBE EL TITULO DE TU PROYECTO AQUI]
    """,

    "pregunta_de_investigacion": """
    [ESCRIBE UNA PREGUNTA ESPECIFICA Y RESPONDIBLE]
    Ejemplo: '¿Cual es el tamano optimo de nanoparticulas de ZnO
    para maximizar la degradacion de azul de metileno bajo luz UV?'
    """,

    "justificacion": """
    [Por que es importante este problema? 2-4 oraciones]
    """,

    "dominio": "A",  # A, B, C, D, o E (ver lista arriba)

    # --- Datos ---
    "fuente_de_datos": """
    [De donde vendran tus datos?]
    Opciones: dataset del curso (nanoparticles_dataset.csv),
              Materials Project API, datos sinteticos generados,
              dataset publico (indicar URL), simulaciones ASE, otro.
    """,

    "n_muestras_estimado": 0,   # Cuantas filas/muestras estimadas

    # --- Herramientas (marca las que usaras) ---
    "herramientas_a_usar": {
        "U1_modelado_atomistico": False,   # ASE, clusters, descriptores
        "U2_simulacion_MD_DFT":    False,   # MD, optimizacion geometrica
        "U3_ml_clasico":           False,   # sklearn, regresion, clasificacion
        "U3_redes_neuronales":     False,   # PyTorch/TF, MLP, CNN
        "U4_llms_generativa":      False,   # LLMs, embeddings, generacion
        "U4_analisis_datos_exp":   False,   # Espectroscopia, imagenes SEM/TEM
        "U5_agentes_langchain":    False,   # LangChain, LangGraph
        "U5_multiagente_crewai":   False,   # CrewAI, multiple agentes
        "U5_rag_memoria":          False,   # RAG, Graph RAG, memoria episodica
        "U6_api_fastapi":          False,   # Despliegue como servicio
    },

    # --- Metodologia ---
    "pasos_del_proyecto": [
        "1. ...",  # Ej: Obtener/generar dataset
        "2. ...",  # Ej: Preprocessing y EDA
        "3. ...",  # Ej: Entrenar modelo
        "4. ...",  # Ej: Optimizacion
        "5. ...",  # Ej: Evaluacion y reporte
    ],

    # --- Resultados esperados ---
    "resultado_principal": """
    [Que producira tu proyecto? Un modelo, un diseno optimo,
     una API, un dataset anotado, etc.]
    """,

    "metrica_de_exito": """
    [Como sabras que tu proyecto funciono bien?]
    Ejemplo: 'R² > 0.80 en prediccion de band gap' o
             'precision > 85% en clasificacion de morfologia'
    """,

    # --- Riesgos ---
    "riesgo_principal": """
    [Cual es el mayor riesgo de tu proyecto y como lo mitigas?]
    Ejemplo: 'Si no hay suficientes datos, generare datos sinteticos
    con el modelo fisico de U2'
    """,
}

print("Propuesta registrada. Ejecuta la siguiente celda para validarla.")
```

```python
# ============================================================
#   VALIDADOR AUTOMATICO DE PROPUESTA
#   Comprueba que la propuesta tiene el minimo requerido
# ============================================================

def validar_propuesta(p: dict) -> tuple[bool, list[str]]:
    """Valida una propuesta de proyecto. Retorna (ok, lista_de_observaciones)."""
    obs = []
    ok = True

    def check(cond, msg_fail, msg_ok=None):
        nonlocal ok
        if not cond:
            obs.append(("FALTA",  msg_fail))
            ok = False
        elif msg_ok:
            obs.append(("OK", msg_ok))

    # Identificacion
    check(p["nombre"] not in ("", "..."), "Campo 'nombre' vacio")

    # Pregunta de investigacion
    pregunta = p["pregunta_de_investigacion"].strip()
    check("ESCRIBE" not in pregunta and len(pregunta) > 30,
          "Pregunta de investigacion no completada o demasiado corta",
          "Pregunta de investigacion presente")

    # Datos
    fuente = p["fuente_de_datos"].strip()
    check("De donde" not in fuente and len(fuente) > 20,
          "Fuente de datos no especificada",
          "Fuente de datos especificada")
    check(p["n_muestras_estimado"] > 0,
          "n_muestras_estimado debe ser > 0")

    # Herramientas: al menos 3 unidades distintas
    herramientas_activas = [k for k, v in p["herramientas_a_usar"].items() if v]
    unidades = set(h.split("_")[0] for h in herramientas_activas)
    check(len(unidades) >= 3,
          f"Solo {len(unidades)} unidad(es) seleccionada(s). Se requieren al menos 3.",
          f"Herramientas de {len(unidades)} unidades distintas")

    # Pasos
    pasos_validos = [p for p in p["pasos_del_proyecto"] if p.strip() not in ("1. ...", "2. ...", "3. ...", "4. ...", "5. ...", "")]
    check(len(pasos_validos) >= 3,
          f"Describe al menos 3 pasos metodologicos (tienes {len(pasos_validos)})",
          f"{len(pasos_validos)} pasos metodologicos definidos")

    # Metrica de exito
    metrica = p["metrica_de_exito"].strip()
    check("Como sabras" not in metrica and len(metrica) > 20,
          "Define una metrica de exito cuantitativa",
          "Metrica de exito definida")

    return ok, obs


print("=" * 55)
print("VALIDACION DE PROPUESTA")
print("=" * 55)

es_valida, observaciones = validar_propuesta(mi_propuesta)

for estado, mensaje in observaciones:
    icono = "[OK]   " if estado == "OK" else "[FALTA]"
    print(f"  {icono} {mensaje}")

print()
if es_valida:
    print("Propuesta LISTA para enviar al instructor.")
    print("Continua con U6_02 para mapear tus herramientas.")
else:
    print("Propuesta INCOMPLETA. Corrige los puntos marcados con [FALTA]")
    print("y vuelve a ejecutar esta celda.")
```
